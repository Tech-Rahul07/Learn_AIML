```python
# learning seaborn
# importing the seaborn lib
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# seaborn has some induilt dataset
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
 tips.head()
```

```python
# visualize the data
sns.set_theme()
sns.relplot(data = tips ,x = 'total_bill',y = 'tip',col = 'time',hue ='smoker',style = 'smoker',size = 'size')
```

```python
titanic  = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# count plot
sns.countplot(x = 'class', data = titanic)

# the no of people is thrid class is more
# second class is lower
# first class is lower then thrid class people

```

```python
titanic.shape
```

```python
# how many people survived
sns.countplot( x =  'survived',data =  titanic)
# not suvived  person is more compare then survived person
```

```python
# bar chart
sns.barplot(data =  titanic, x = 'sex',y ='survived',hue =  'class')
```

```python
# now importing house pricre dataste
boston_df = pd.read_csv('BostonHousing.csv')
```

```python
boston_df.head(100)
```

```python
# distribution plot
sns.displot(boston_df['rm'])
sns.distplot(boston_df['rm'])
```

```python
print(boston_df['rm'])
```

```python
# correlation matrix
correlaton = boston_df.corr()
```

```python
# correlation by heatmap

# constrating heatmap
plt.figure(figsize = (10,10))
sns.heatmap(correlaton,cbar = True,square = True,fmt = '.1f', annot =True,annot_kws = {'size':8},cmap = 'Blues')
```

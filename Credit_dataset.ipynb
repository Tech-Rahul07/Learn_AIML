```python
# chek the distribution of two class
data['Class'].value_counts()
```

```python
# this is imbalece dataset here 0 is normal transactions
# and 1 is fraud transactions
legit = data[data.Class == 0]
froud = data[data.Class == 1]
```

```python
legit.shape
```

```python
froud.shape
```

```python
# use undersampling method
# build the sample dataset containing similer distribution of legit and froud trasnstion
# number of froud trans -->492

legit_sample = legit.sample(n = 492)
```

```python
legit_sample.shape
```

```python
# concating of ligit_sample and froud data frame
new_dataset = pd.concat([legit_sample,froud],axis = 0)
```
